### Set Methods

```python
#Eleement adding method
#add(ele): it adds the element to set if elememnt not in set,otherwise nothing
s = {1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
s.add("Code")
print(s)
```

# removing methods
# remove(item): it removes the item if it exists in set operations otherwise 'key error'
# discard (item): it removes the item if it exists in set operations otherwise 


```python
s = {1,2,3,4,5,4,3}
print(s)
#s.remove(100)  #Key Value Error
s.discard(3)
print(s)
s.discard(100) 
print(s)
```

### SET OPERATIONS

```python
# union(set) : it retrns all unique elements from all sets
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
union =s1.union(s2,s3)
print(union)
res = s1 | s2| s3
print(res)
```

```python
s1 = {1,8,5,7}
s2 = {1,2,5,6,7}
s3 = {1,2,8,9}
i= s1.intersection(s2,s3)
print(i)
res = s1&s2&s3
print(res)
```

```python
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
diff =s1.difference(s2,s3) # this means it returns the elements that are not in s2 and s3
print(diff)
res =(s1 |s2)- s3
print(res)
```

```python
#set operations
#symmetric_difference(set): it retyrns non common elements
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {5,1,2,8,9}
symm =s1.symmetric_difference(s3) # this means it retuens the elements that are not in s2 and s3
print(symm)
res=s1 ^ s2
```

```python
#issubset(set)
#issuperset(set)
#isdisjoint(set)
s1={5,7}
s2={1,5,7}
subset=s1.issubset(s2)
print(subset)
```

```python
s1={4,7}
s2={4,1,5,7}
superset = s2.issuperset(s1)
print(superset)
```

```python
s1={3,2}
s2={1,5,7}
dis=s1.isdisjoint(s2)
print(dis)
```
