```python
s1=input()
s2=input()
s=''
for i in range(len(s1)):
    diff=abs(ord(s1[i])-ord(s2[i]))
    if diff == 0:
        s+=s1[i]
    else:
        s += chr(64+diff)
print(s)
```

```python
s1=input()
s2=input()
b=''
s=''
for i in range (len(s1)):
    if ord(s1[i])>ord(s2[i]):
        b+=s1[i]
        s+=s2[i]
    elif ord(s2[i])>ord(s1[i]):
        s+=s2[i]
        b+=s1[i]
print(b)
print(s)
```

```python
# map(fun, iterable): It applies the functionaliy on eah element in iterable
# filter(fun, iterable): It filters the iterables, it returns the values which is not None
# find the square of list of integers
def square(x):
    return x ** 2
lst = [1,2,3,4,5]
print(list(map(square,lst)))
```

```python
# sum of the list of tuples (We can pass multiple values)
def even(t):
    x, y = t
    return x + y
arr = [(1,2),(3,4),(5,6)]
print(list(map(even,arr)))
```

```python
# filter the evnen numbers from list of integers
def even(x):
    if x % 2 == 0:
        return x
arr = [1,2,3,4,5,6,7]
print(list(filter(even,arr)))
print(list(map(even,arr)))
```

```python
#zip (iterable1, iterable 2,.........., iterableN)
#It return list of tuples ,the tuple is generated as each position elements from all iterables
name = ['ravi','srinu','siva']
age=[20,21,22]
print(list(zip(name,age)))
```

```python
#enumerate(iterable):it returns the list of tuple of index and value
arr = [10,20,30]
print(list(enumerate(arr)))
```

```python
#find the target element position 
arr=[10,20,30,40,50]
target=40
for i,j in enumerate(arr):
    if target==j:
        print(i)
        break
else:
    print(-1)
        
```

```python
# count the frequency of characters in string.
string = "babu"
dic = dict()
for i in string:
    dic[i] = string.count(i)
print(dic)
print(list(dic.values()))
```
