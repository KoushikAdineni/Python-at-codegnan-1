### Set Characteristics


```python
#set unordered collection
#set returns unique elements
s={1,2,3,2,3,'hi','a',2.02, (10,20,30)}
print(s)
#print(s[0]) # set doesnot support indexing
```

```python
#list cant be used as set element

s={1,2,[3,2,3],'hi','a',2.02, (10,20,30)}

print(s)
```

```python
#empty set
s1 ={ }
s2=()
print(type(s1),type(s2))
```
