```python
#scope of variables
#1 global scope
#2 local scope
#3 enclosing scope
#4 builtin scope
x =10
def outer():
    y=20
    def inner():
        z=30
        print(z,x,y)
    inner()
outer()
```

```python
#using global scope variable in local scope
x=100
def outer():
    y=x+1
    print("local scope variable:",y)
outer()
print("global scope variable:",x)
```

```python
#updating the global scope variable value in local scope
#cant update global scope variable 
#updating the global scope variable value in loacal scope by using 'global' keyword
x=100
def outer():
    x=x+1
    print(x)
outer()
print(x)
```

```python
#updating the global scope variable value in loacal scope by using 'global' keyword
x=100
def outer():
    global x
    x=x+1
    print(x)
outer()
print(x)
```

```python
#updating the enclosure scope variable value in local scope
x =10 #global scope varoable
def outer():
    y=20 #enclosing scope variable
    def inner():
        z=30
        y=y+z#local scope variable function
        print(z,x,y)#built in scope
    inner()
    print("enclosing scope variable :",y)
outer()
```

```python
#updating the enclosure scope variable value in local scope
x =10 #global scope varoable
def outer():
    y=20 #enclosing scope variable
    def inner():
        nonlocal y
        z=30
        y=y+z#local scope variable function
        print(z,x,y)#built in scope
    inner()
    print("enclosing scope variable :",y)
outer()
```

```python
#if the value changes in local that does not change the value in global
x =100
def outer():
    x=200
    x=x+2
    print("local scope value of x:",x)
outer()
print("global scope value of x:",x)
```
