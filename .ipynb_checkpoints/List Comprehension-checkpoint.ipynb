```python
#generate 1-n numbers using list comprehension
n= int(input())
res = [x for x in range(1, n+1)]
print(res)
```

```python
#first n even numbers using list comprehension
n= int(input())
res = [x*2 for x in range(0,n)]
print(res)
```

```python
n= int(input())
res = [x**3 for x in range(1,n+1)]
print(res)
```

```python
#find the odd numbers between n1 to n2
n1,n2 = map(int,input().split())
res = [x for x in range(n1,n2+1)]
print(res)
```

```python
arr = list(map(int,input().split()))
res = [x**2 if x%2==0 else round(x**0.5,2) for x in arr]
print(res)
```

```python
#2 dices combination values
res =[(x,y) for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
#sum of the 2 dices is divisible by 3
res =[(x,y) for x in range(1,7) for y in range(1,7) if (x+y)%3 == 0]
print(res)
```

```python
#return any dice value as 4
res =[(x,y) for x in range(1,7) for y in range(1,7) if x== 4 or y==4]
print(res)
```

```python
#count the number of vowels exists in string
string = input()
res =[char for char in string.lower() if char in"aeiou"]
print(res)
print(len(res))
```

```python
n=int(input())
res = {x for x in range(1,n+1)}
print(res)
print(type(res))
```

```python
n=int(input())
res = {x:x**2 for x in range(1,n+1)}
print(res)
print(type(res))
```

```python
n=int(input())
res = tuple((x**2) for x in range(1,n+1))
print(res)
print(type(res))
```

```python
def reverse_string(s):
    s=list(s)
    i=0
    j=len(s)-1

    while i<j:
        temp = s[i]
        s[i] = s[j]
        s[j] = temp

        i+=1
        j-=1                               #wrong

        return "".join(s)
str = input()
print(reverse_string(str))
        
```
