```python
#find even or odd among 3 numbers
n1,n2,n3=map(int,input().split())
if n1%2==0:
    print(f"{n1} is even")
else:
    print(f"{n1} is odd")
if n2%2==0:
    print(f"{n2} is even")
else:
    print(f"{n2} is odd")
if n3%2==0:
    print(f"{n3} is even")
else:
    print(f"{n3} is odd")
    
```

```python
lst = map(int,input().split())
for num in (lst):
    if num%2==0:
        sq=num**2
        print(sq)
    else:
        add=num+1
        print(add)
        
```

```python
lst = map(int,input().split())
for num in lst:
    if num%3==0 and num%5==0:
        print("fizzbuzz")
    elif num%3==0:
        print("fizz")
    elif num%5==0:
        print("buzz")
    else:
        print(num)
```

```python
word = input()
v=0
c=0
vowels="aeiouAEIOU"
for ch in word:
    if ch in vowels:
        v+=1
    else:
        c+=1
print(v)
print(c)
```

```python
word = input()
v=0
c=0
vowels="aeiouAEIOU"
for ch in word:
    if ch.isalpha():
         if ch in vowels:
             v+=1
        
         else:
              c+=1
print(v)
print(c)
        
   
```

```python
#loop control statements
lst=[11,12,9,5,6,7,3]
target =5
n=len(lst)
for i in range(n):
    if lst[i]==target:
        print(i)
        break
else:
    print("-1")
```

```python
#continue
for i in range(1,11):
    if i%2==0:
        continue   #skips the answer and goes to the next iteration
    print(i)
```

```python
lst=list(map(int,input().split()))
tar=int(input())
n=len(lst)
for i in range(n):
    if lst[i]==tar:
        print(i)
        break
else:
    print("-1")
```

```python
n=int(input())
fact =list()
for i in range (1,n+1):
    if n%i==0:
        fact.append(i)
print(fact)
if len(fact)==2:
    print(f"{n} is prime")
        
```

```python
n=5
k=1
for i in range(n):
    for j in range(n):
        if i%2==0:
            print(i*n+j+1,end="\t")
            k+=1
        else:
            print((i+1)*n-j,end="\t")
            k-=1
        
    print()
        
```

```python
n=5
k=n*(n+1)//2
for i in range(n):
    for j in range(n):
        if i>=j:
            print(k,end="\t")
            k-=1
        else:
            print(" ",end="\t")
    print()
    
```

```python
# hallow right angle triangle
n=5

for i in range(n):
    for j in range(n):
        if j==0 or i==0 or j==n-i-1:
            print(chr(65+j),end=" ")
            
        else:
            print(" ",end=" ")
    print()
```

```python
n=5
k=0
for i in range(n):
    for j in range(n):
        if i>=j:
            print(chr(65+k),end=" ")
            k=k+1
        else:
            print(" ",end=" ")
    print()
```
