## Polymorphism:
- In Python polymorphism, Poly Means 'Many' and Morphism means 'forms'.
- In Polymorphism single function , it shows different behaviours based on input context(values).
- In Python polymorphism can achieve through the following ways
1. through operators
2. polymorphism through functions
3. polymorphism through Methods
   3.1 Runtime Polymorphism (Method overriding)
   3.2 Compile time  Polymorphism (Method overloading)
### 1.polymorphism through operators:
- Here Single operator it shows different actions or behaviours based on input values
- for example: '+' Operator it performs addition operation when inputs are integers as well as it performs string concatenation when inputs are strings.
- 3+5=8 (addition'+')
- code + gnan (string concatenation'+')
- [1,2,3] + [4,5] (list concatenation'+')
### 2.polymorphism through functions:
- here a function shows different actions based on input context
- ex: len('Hello') ---> string length
- len([10,2,3,5]) ----> List length
### 3.polymorphism through Methods:
1. Run Time Polymorphism(method Overriding): A method implemented in child class or defined in child class that is already exist in parent class with the same name whenever if you call the method through the child class object,the child class method over rides the parent class method
- class A
-    def show()  #same method
- class B(A):
-    def show() #same method
- b = B()
- b.show() # calling same method from child class only
2. Complie time Polymorphism (Method overloading): A  class contains multiple methods with same name with diffreent parameters list,in that case last defined last defined method override the previoues defined method
  
- directly python doesnot support method overloading but we can achieve same functionality through default arguments, variables ,length postional,or variable length key word arguments
- class 
  

```python
class person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and my age and gender is {self.age} , {self.gender}"
        
class student(person):
    def __init__(self,name,age,gender,stdid,branch):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.branch = branch

    def details(self):
        return f"My name is {self.name} and my stid is {self.stdid} and branch is {self.branch}"

s = student("koushik",22,"male",101,"ai")
print(s.details())
#here as we can see that the  child class overrides the parent class
```

```python
class person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and my age and gender is {self.age} , {self.gender}"
        
class student(person):
    def __init__(self,name,age,gender,stdid,branch):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.branch = branch

    def details(self):
        print(super().details())  # if we want to print the parent details , we can use the super(). fun that takes the details from parent class
        return f"My name is {self.name} and my stid is {self.stdid} and branch is {self.branch}"

s = student("koushik",22,"male",101,"ai")
print(s.details())
#here as we can see that the  child class overrides the parent class
```

```python
# method overloading
class Math:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
    def add(self,a,b,c,d):
        return a+b+c+d

b=Math() 
#print(b.add(2,3)) #missing 2 required positional arguments: 'c' and 'd'
#print(b.add(2,3,4)) #missing 1 required positional argument: 'd'
print(b.add(1,2,3,4)) # only print of last fun based on arguments
```

```python
class math:
    def add(self,a,b,c=0,d=0):
        return a+b+c+d

b = math()
print(b.add(2,3)) 
print(b.add(2,3,4)) 
print(b.add(1,2,3,4))
# python directly doesnot support overloading ,but we can achieve same fun through the default arguments or arbitary arguments
```

```python
# __str__() magic methods
class point:
    def __init__(self,x):
        self.x = x
    def __str__(self):
        return f"this is point class and point value is {self.x}"
    def __add__(self,other):
        return self.x + other.x
p1=point(5)
print(p1)
p2=point(10)
print(p1+p2)
print(p1.x+p2.x)
p3=p1+p2
print(p3)
```

```python
class Point:
    def __init__(self,x):
        self.x=x
    def __add__(self,other):
        return self.x + other.x
    def __sub__(self,other):
        return self.x - other.x
    def __mul__(self,other):
        return self.x * other.x
    def __truediv__(self,other): # truediv for division
        return self.x /other.x
    def __mod__(self,other):
        return self.x % other.x
p1=Point(5)
print(p1)
p2=Point(4)
print(p2)
print(p1+p2)
print(p1-p2)
print(p1* p2)
print(p1/p2)
print(p1%p2)
```
