###Methods:The Funtion definition is implemented inside the class those functions called as Methods
-Types of Methods
1. Instance Method
2. class Method
3. Static Method


#### Instance Method :  
- The method defination take 1st parameter as "self" those methods called as 'Instance Method'. This method works on instance attributes.
   - syntax:
     -  class class_name:           <br>
        - def method_name(self, parameters):     <br>
            -  block of code         <br>

#### Class Method :
  - The Method defination starts with 'classmethod' decorator and it takes 1st parameter as 'cls' those methods called as 'Class methods'. This methods works on class attributes.
  - syntax:
      -  class class_name:          <br>
         - @classmethod             <br>
        - def method_name(cls, parameters):     <br>
             -  block of code        <br>
    

#### static method:
- This Method defination starts with 'staticmethod' decorator and it doesnot take 1st parameter as 'self or cls' those methods called as 'Static methods'. This methods works on except instance attributes and class attributes.
   - syntax:
      -  class class_name:          <br>
         - @staticmethod             <br>
        - def method_name(parameter):    <br>
            -  block of code        <br>

### Encapsulation:
- Encapsulation means, wraping the methods and attributes in single unit that one is class
       the main aim of encapsulation is to restrict the direct access of attributes and methods
- This type of restriction can implement by using 'Access Modifiers'
### Types of Access modifiers:
1. public
2. protected
3. private
- In python we dont have any keywords to represent it public,  protected and private,
 but by using underscore(_) can define.
### 1.public : 
- The Public Members(i.e., attributes + method) doesnt starts with underscore(_)
  and this public members can access any where in the program.
### 2. Protected : 
- The Protected Members (i.e., attributes + method) starts with single underscore(_)
  and this members can access inside the class and in subclass also.
### 3. Private : 
- The Private Members (i.e., attributes + method) starts with double underscore(_)
  and this members can access only inside the class.
  

```python
class Bank:
    def __init__(self, name, account, password, balance):
        self.name = name
        self.account = account
        self.password = password
        self._balance = balance      # Protected
        self.__totalusers = 101      # Private

    def get_balance(self):
        return self._balance

    def deposit(self, amount):
        self._balance += amount

    def __show_total_users(self):    # Private Method
        return self.__totalusers


u1 = Bank("Ravi", 1001, "Ravi@123", 10000)

print(u1.name)
print(u1._balance)                     # Protected access
print(u1._Bank__totalusers)            # Name Mangling (private variable)
print(u1._Bank__show_total_users())    # Name Mangling (private method)
# java is safer in protection 
# to access the protected memeber (obj._protected member name)
#to access the private members by using name mangling(obj._classname__private member name)
# the advantage of encapsulation is we can provide the security for data as well as for functionalites
```
