```python
#define list 
#list is ordered collection
lst = [1,2,3,4,"Hi","Hello",1,3,4]
print(lst)
```

```python
#list allows inexing and slicing
#accessing list items
lst=[1,2,3,4,"Hi","Hello",1,3,4]
print(lst[5])
print(lst[-1])
```

```python
lst =[10,20,30,40,50,60,70]
print(lst[:5:-1])
print(lst[::100])
print(lst[::-1])
```

```python
#list operations
#1.list concatination (using + operator)
#2.list repetation(using *operator)
l1=[1,2,3,4]
l2=[3,4,5,6,7]
print(l1+12)
print(l1+l2)
print(l1*3)
print(0*10)
```

```python
List Methods
Item adding methods
append(item) : It adds item at end of list
extend (iterable) : It adds the all iterator elements at end of list
insert(pos, item) : It adds item at specified position
```

```python
lst= [10,20,30,40]
lst.extend([1,2,3])
print(lst)
lst.append([100,200])
print(lst)
lst.insert(0,111)
print(lst)
```

```python
#item removing methods
#remove(item):it removes the first occurance item from the list if it exits in list otherwise Valueerror
#pop(pos):
```

```python
lst = [10,20,30,40]
lst.remove(20)
print(lst)
lst.remove(100)#value error
print(lst)
```

```python
lst = [10,20,30,40,50,60]
val=lst.pop()
print(val)
print(lst)
val2=lst.pop(1)#it removes the specific element
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
#index and count
#index(item):if item is presence it returns the occurence the index otherwise Value Error
#count(item):it return the count of total occurence of item
lst =[10,20,30,40,20,20,50]
val=lst.index(20)
print(val)
count=lst.count(20)
print(count)
```

Sorting items
sort(reverse = False): By default it sorts the origing list in into ascending order. if reverser is True it sorts in Descending order

```python
lst= [1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```

```python
lst= ['hello',"hi"]
lst.sort()
print(lst)
lst.sort(reverse=True)#only single datatype list sort support
print(lst)
```
