##String Methods
- case conversion Methods
   - upper() :It converts entire string into upper case
   - lower() :It converts entire string into lower case
   - title() :It converts the first letter of every word into upper and remining to lower
   - swapcase() :It converts upper case letter into lower case and Lower to upper
   - captilize() :It conerts First character into upper case and remaining into lower case

```python
string = "The sky is blue and Ocean is blue"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

- Remove white space or pattern at ends of string
      -strip() : By default it moves the white space at both ends string
      -lstrip():it removes the white space at left side of string
      -rstrip(): it removes the white space at right side of string

```python
s="@@Python@@class@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
s="      Pythonclass      "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
#split(): It splits the entire string into list of 
# split(): It splits the entire string into list of substring based on specified separetor
# by default it splicts at space 
string= "The Sky is blue and Ocean is blue"
print(string.split())
print(string.split("is"))
print(string.split("e"))
```

```python
#'sep'.join(iterable): It joins the all iterable elements with seperator
words = ['The', 'Sky', 'is', 'blue', 'and', 'Ocean', 'is', 'blue']
print("".join(words))
print(" ".join(words))
print("-".join(words))
```

```python
# count (substring): it returns the total non-overlapping dccurences of substring 
s ="ABCBCBCaabb"
print(s.count("B"))
print(s.count("BCB"))
lower_string = s.lower()
print(lower_string)
print(lower_string.count('b'))
upper_string=s.upper()
print(upper_string)
print(upper_string.count('B'))
```

###String checking Methods
- string checking methods returns the boolean values as result
-is upper():it returns true ,if entire string is in upper case
-is lower():it returns true ,if entire string is in lower case
-isalpha():it returns true ,if entire string contains alphabets
-isdigit():it returns true ,if entire string contains digits
-isalnum():it returns true ,if entire string contains either alphabets or digits
-startwith(substring): it returns true if,string starts with substring 
-endwith(substring):it returns true ,if string ends with substru

```python
print("CODE".isupper())
print("CODE".islower())
print("1234".isdigit())
print("COde".isalpha())
print("CO4E".isalnum())
print("code".startswith('c'))
print("CODE".endswith('e'))
```
