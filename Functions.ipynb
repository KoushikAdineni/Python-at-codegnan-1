```python
#Types of functions based on parameters and return values
#4. with parameters and  with return
# function definition block
def addition(x:int, y:int)->int:
    c = x + y
    return c
#function calling block
a , b =10, 20
res = addition(a,b)
print(res)
```

```python
#3 with paramters and without return
#by default function returns none value,when function not returning any value
def addition(x:int, y:int)->int:
    c = x + y
    print(f"This is from function definition block:{c}")

#function calling block
a , b =10, 20
res = addition(a,b)
print(f"This is from function calling block:{res}" )
```

```python
#2 without paramters and with return
def addition()->int:
    x = 10
    y = 20
    z = x+y
    return z
#function calling block
res = addition()
print(f"This is from function calling block:{res}" )
```

```python
#1 with out parameter and without return
def addition() ->None:
    x = 10
    y = 20
    c = x+y
    return c
    print(f"This is from function definition block:{c}" )
#function calling block
res = addition()
print(f"This is from function calling block:{res}" )
```
