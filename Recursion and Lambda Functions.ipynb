```python
#Direct recursion
def fun(x):
    if x == 0:
        return 0
    x = x-1
    fun(x)
    print(x)
fun(3)
```

```python
#direct recursion directly in function
def fun(x):
    if x == 0:
        return 0
    fun(x-1)
    print(x)
fun(3)
```

```python
#factorial using recursion
def fact(x):
    if x==0 or x==1:
        return 1
    return x*fact(x-1)
fact(5)
```

```python
# find sum of the digits in a number
def sum(x):
    if x==0:
        return 0
    return x % 10 + sum(x//10)
print(sum(456))
print(sum(15346))
```

```python
# count the digits in a number

def count_digits(x):
    if x==0:
        return 0
    return 1 + count_digits(x//10)
print(count_digits(456))
print(count_digits(45678))
```

```python
# find the squares of list of integers using map and lambda function
arr = [1,2,3,4,5,6]
print(list(map(lambda x : x**2,arr)))
```

```python
arr = [1,2,3,4,5,6]
print(list(filter(lambda x : x%2==0,arr)))
```
