```python
#generate total bill morethan 500 give 10% discount else no or 0% discount
total=int(input("enter the amount:"))
if total>500:
    dis = total*10/100
    dis_bill=total-dis
    print(f"{dis_bill} is your bill")
else:
    print(f"{total} is your bill")
```

```python
## 5.Question
products={'pen':5,'book':20,'pencil':3}
item=input().lower()
quantity=int(input())
if item in products:
    total_amount=products.get(item)* quantity
    print(f'Total amount of {quantity} {item} is: {total_amount}')
else:
    print(f"Item not avaliable")
```

```python
#2 case
n1,n2,n3 =map(int,input().split())
if n1>n2 and n1>n3:
    print(f"{n1} is bigger")
elif n2>n3 and n2>n1:
    print(f"{n2} is bigger")
else:
    print(f"{n3} is bigger")
```

```python
### find the biggest number among the three numbers(using nested if)
n1,n2,n3 =map(int,input().split())
if n1>n2:
    if n1>n3:
        print(f"{n1} is bigger")
    else:
        print(f"{n3} is bigger")
else:
    if n2>n3:
        print(f"{n2} is bigger")
    else:
        print(f"{n3} is bigger")
        
```
