```python
n=int(input())
count=0
for i in range(2,n+1):
    if i%2==0:
        print(i)
        count+=1

if count==0:
        print("-1")
```

```python
n=int(input())
num=list(map(int,input().split()))
Even=0
Odd=0
Positive=0
Negative=0
for i in num:
    if i%2==0:
        Even+=1
    elif i%2==1:
        Odd+=1
    elif i>0:
        Positive+=1
    else:
        Negative+=1
print('Even:',Even)
print('Odd:',Odd)
print('Positive:',Positive)
print('Negative:',Negative)
```
