```python
def fun(n):
    res=[]
    for i in range(1,n+1):
        res.append(i)
    return res
print(fun(10))
        
```

```python
# 1 to n numbers using functions
import sys # sys used to get the size of system varible
def numbers(n):
    res=[]
    for i in range(1,n+1):
        res.append(i)
    return res
r=numbers(10)
print(sys.getsizeof(r))
print(r)
```

```python
import sys # sys used to get the size of system varible
def gen(n):
    res=[]
    for i in range(1,n+1):
        yield i
    return res
r=gen(3)
print(r)
n=next(r)
print(sys.getsizeof(n))
print(n)
print(next(r))
print(next(r))
print(next(r)) # StopIteration: [], Because there is no element left to get value in generator object
```

```python
import sys # sys used to get the size of system varible
def gen(n):
    res=[]
    for i in range(1,n+1):
        yield i
    return res
r=gen(3)
print(r)
# to prevent the stopIteration error by using for loop
for val in r:
    print(val)
```

```python
#first thousand cubic values using functions
import time
s = time.time()
def fun(n):
    res=[]
    for i in range(1,n+1):
        res.append(i**3)
    return res
print(fun(100))
e =time.time()
print(e-s)
        
```
