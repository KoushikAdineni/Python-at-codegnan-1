Dictionary Methods
 - get(key, default_value:Optional):it returns the value if key is exists in dictionary otherwise it returns default value

```python
# get method
#dictionary keys are unique
details={"name":"Shreyas","course":"Python","batch":57,"age":22}
print(details.get('name'))
print(details.get('marks')) # it returns none,by default bcz key not exist in details
print(details.get('marks','marks not avaliable')) ##it returns default value.
```

Update Method
-update(new_dictionary): This method updates the values if new_dictionary keys already exists in Dictionary , otherwise it adds new items todictionary

```python
details={"name":"Shreyas","course":"Python","batch":57,"age":22}
details.update({"name":"kohili","course":"cricket","location":"delhi"})
print(details)
```

-keys: This method returns the list dictionary keys 

-values: This method returns the list dictionary Values

-items:This Method return the list of Tuple of Key-value pairs

```python
details={'name': 'kohili', 'course': 'cricket', 'batch': 57, 'age': 22, 'location': 'delhi'}
print(details.keys())
print(details.values())
print(details.items())

```

Item Removing Methods
- popitem():It return and removes the last item as tuple
- pop(key):This method removes the specified key item and return the specific key's value,otherwise KeyError

```python
details={"name":"Shreyas","course":"Python","batch":57,"age":22}
item=details.popitem()
print(item)
print(details)
val=details.pop('name')
print(val)
print(details)
```
