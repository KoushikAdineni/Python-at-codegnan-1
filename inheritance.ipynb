#### Inhertance:
- In python , a class gets properties from another class this process is called as inheritence
- a child class is inherited from parent class
## Types:
1. single
2. multilevel
3. multiple
4. hierachial
5. hybrid
## 1.Single Inhertance:
  (single parent-------->single child)
- A child class get properties from single parent class.This is called as "single Inheritance"
    syntax:
            class parent_class_name:
                  #attributes
                  #method
            class child class_name(parent_class_name):
                   #attributes
                   #methods
## 2.Multi Level Inheritance:
    (Grand Parent ---->Parent------>child)
- A Class (child) gets properties from some another class or another class(parent) that class also get propeties from
  another class(Grand parent)
  eg:- mobile---->oppo----->reno 10
     - animal---->dog------>puppy
     - employee--->TL------->Manager-------->HR
## 3.Multiple Inheritance:
- A Single child class get proprties from multiple parent classs this is called mulitple inheritance.
     (mother,father------->child)
## 4.Hierachial Inheritance:
- Multiple Parent classes get properties from single parent class.
              (parent------>c1,c2,c3)
  eg: network----->aitel,jio...
## 5.Hybrid Inheritance:
- Combination of two or more types of inheritances.
  
                
  
  









  

```python
#Types of inheritance
#single inheritance
#with out child class implementation
#when child class dont have constructor, then parent class constructor behvaes like child class constructor.
class person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and i am {self.age} years old"
class student(person):
    pass
s=student("Bhanu",20,"Female")
print(s.name)
print(s.details())
```

```python
#Types of inheritance
#single inheritance
# child class with constructor
#when child class dont have constructor, then parent class constructor behvaes like child class constructor.
class person:
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and i am {self.age} years old"
class student(person):
    def __init__(self, stdid,name,age,branch):
        super().__init__(name,age,gender)
        self.stdid = stdid
        self.branch = branch
        
        
s=student(101,"Bhanu","female",20,"AI")
print(s.name)
print(s.branch)
print(s.details())
#print(s.gender)
```
