### TYPE CONVERSIONS:
 - Converting one data type format into another data type format this is called as type conversion
 - they are 2 types:
 - 1) Implicit type conversion
   2) explicit type conversion(Type casting)
 
1) IMPLICIT TYPE CONVERSION:
 - The python interpreter automatically coverts the datatype becz it prevents the data loss
 - Implicit type conversion can happens in arthimetic operators

2)EXPLICIT TYPE CONVERSION OR TYPE CASTING:
 - here type conversion happens using python built in  datatype functions

### python built in  datatype functions:
  1.Integer- int()
  2.Float value- float()
  3.Complex value- complex()
  4.Boolean- bool()
  5.String - str()
  6.List- list()
  7.Tuple - tuple()
  8.Set- set()
  9.Dictionary - dict()

```python


# integer to float, boolean, string
num=10
print(float(num))
print(bool(num))
print(str(num))
print(complex(num))
```
