```python
name = input("enter name:")
age = int(input())
print("my name is",name,"and my age is",age)
print("my name is "+ name +" and my age is "+ str(age))
```

```python
#using access specifiers
#string -- %s
#integer -- %d
#float -- % f
name = "koushik"
age=22
print("my name is %s and my age is %d"%(name,age))
```

```python
name = "koushik"
age=22
print("my name is %s and my age is %d"%(age,name))##checking the possiblity learn it 
```

```python
#dot format
name = "koushik"
age=22
print("my name is {} and my age is {}".format(name,age))
print("my name is {} and my age is {}".format(age,name))#its possible in format type
```

```python
#f format
name = "koushik"
age=22
print(f"my name is {name} and my age is {age}")
```
