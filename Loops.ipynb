```python
#find even or odd among 3 numbers
n1,n2,n3=map(int,input().split())
if n1%2==0:
    print(f"{n1} is even")
else:
    print(f"{n1} is odd")
if n2%2==0:
    print(f"{n2} is even")
else:
    print(f"{n2} is odd")
if n3%2==0:
    print(f"{n3} is even")
else:
    print(f"{n3} is odd")
    
```

```python
lst = map(int,input().split())
for num in lst:
    if num%3==0 and num%5==0:
        print("fizzbuzz")
    elif num%3==0:
        print("fizz")
    elif num%5==0:
        print("buzz")
    else:
        print(num)
```

```python
word = input()
v=0
c=0
vowels="aeiouAEIOU"
for ch in word:
    if ch in vowels:
        v+=1
    else:
        c+=1
print(v)
print(c)
```

```python
#loop control statements
lst=[11,12,9,5,6,7,3]
target =5
n=len(lst)
for i in range(n):
    if lst[i]==target:
        print(i)
        break
else:
    print("-1")
```

```python
#continue
for i in range(1,11):
    if i%2==0:
        continue   #skips the answer and goes to the next iteration
    print(i)
```

```python
n=int(input())
fact =list()
for i in range (1,n+1):
    if n%i==0:
        fact.append(i)
if len(fact)==2:
    print(f"{n} is prime")
        
```
